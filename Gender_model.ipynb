```python
# استيراد المكتبات الضرورية
import os
import cv2
from PIL import Image
import imagehash

# مسارات المجلدات
women_path = "/root/.cache/kagglehub/datasets/playlist/men-women-classification/versions/3/women"
men_path = "/root/.cache/kagglehub/datasets/playlist/men-women-classification/versions/3/men"
```

```python
print("Processing Men folder...")

men_image_files = [os.path.join(men_path, f) for f in os.listdir(men_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
print(f"Found {len(men_image_files)} images in Men folder")

for img_path in men_image_files:
    try:
        img = cv2.imread(img_path)
        if img is None or img.shape[0] == 0 or img.shape[1] == 0:
            print(f"Invalid or empty image: {img_path}")
            os.remove(img_path)
            continue
    except Exception as e:
        print(f"Error processing image {img_path}: {e}")
        os.remove(img_path)
```

```python
hash_dict_men = {}

for img_path in men_image_files:
    try:
        img_hash = imagehash.average_hash(Image.open(img_path))
        if img_hash in hash_dict_men:
            print(f"Duplicate image found: {img_path}")
            os.remove(img_path)
        else:
            hash_dict_men[img_hash] = img_path
    except Exception as e:
        print(f"Error processing image {img_path}: {e}")
        os.remove(img_path)
```

```python
print("Remaining images in Men folder:")
remaining_men_images = [f for f in os.listdir(men_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
print(len(remaining_men_images)) # Use len() to get the total number of items in the list
```

```python
import tensorflow as tf
import os
import numpy as np
from tensorflow.keras.preprocessing import image
from sklearn.model_selection import train_test_split

men_dir = '/root/.cache/kagglehub/datasets/playlist/men-women-classification/versions/3/men/'
women_dir = '/root/.cache/kagglehub/datasets/playlist/men-women-classification/versions/3/women/'

def load_images_from_folder(folder, label, img_size=(96, 96)):
    images = []
    labels = []
    for filename in os.listdir(folder):
        img_path = os.path.join(folder, filename)
        img = image.load_img(img_path, target_size=img_size)
        img_array = image.img_to_array(img)
        images.append(img_array)
        labels.append(label)
    return np.array(images), np.array(labels)

men_images, men_labels = load_images_from_folder(men_dir, 0)

women_images, women_labels = load_images_from_folder(women_dir, 1)


all_images = np.concatenate((men_images, women_images), axis=0)
all_labels = np.concatenate((men_labels, women_labels), axis=0)

trainX, testX, trainY, testY = train_test_split(all_images, all_labels, test_size=0.2, random_state=42)
```

```python
trainX = trainX / 255.0
testX = testX / 255.0

trainY = tf.one_hot(trainY, 2)
testY = tf.one_hot(testY, 2)

train_dataset = tf.data.Dataset.from_tensor_slices((trainX, trainY))
test_dataset = tf.data.Dataset.from_tensor_slices((testX, testY))

train_dataset = train_dataset.shuffle(buffer_size=1000).batch(32)
test_dataset = test_dataset.batch(32)
```
